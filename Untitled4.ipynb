```python
# Step 1: Import all required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Make plots look better
plt.style.use("seaborn-v0_8")

print("✅ Libraries imported successfully")
```

```python
# Step 2: Upload the dataset file from your local system

from google.colab import files

uploaded = files.upload()  # Choose: AQI-and-Lat-Long-of-Countries.csv

# After upload, you can see the filenames:
uploaded.keys()
```

```python
# Step 3: Load the dataset into a pandas DataFrame

# Change the filename here if your file name is different
csv_file_name = "AQI-and-Lat-Long-of-Countries.csv"

data = pd.read_csv(csv_file_name)

print("✅ Data loaded successfully!")
print("🔹 First 5 rows:")
display(data.head())
```

```python
# Check basic information about the dataset

print("🔍 Data Info:")
print(data.info())

print("\n🔍 Missing values per column:")
print(data.isnull().sum())

print("\n🔍 Summary statistics:")
display(data.describe())
```

```python
# Step 4: Clean column names (optional but recommended)

print("📝 Original columns:")
print(list(data.columns))

# Strip spaces and make all lowercase with underscores
data.columns = [col.strip().lower().replace(" ", "_") for col in data.columns]

print("\n✅ Cleaned columns:")
print(list(data.columns))
```

```python
# Step 5: Exploratory Data Analysis - Correlation Heatmap

plt.figure(figsize=(8, 6))
corr = data.corr(numeric_only=True)  # numeric_only avoids warnings
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Features")
plt.tight_layout()
plt.show()
```

```python
# Step 6: Define features (X) and target (y)

# Print columns again to be sure
print("Available columns:", list(data.columns))

# Adjust names here if your file has slightly different names
feature_cols = [
    "co_aqi_value",
    "ozone_aqi_value",
    "no2_aqi_value",
    "pm2.5_aqi_value"  # if your cleaned column is `pm2_5_aqi_value`, change this line
    # You can also add 'lat', 'lng' if you want:
    # "lat",
    # "lng"
]

target_col = "aqi_value"

# Safety check: raise error if column not found
for col in feature_cols + [target_col]:
    if col not in data.columns:
        raise KeyError(f"Column '{col}' not found in dataset. Check the column names above.")

X = data[feature_cols]
y = data[target_col]

print("\n✅ Features and target selected")
print("Features:", feature_cols)
print("Target  :", target_col)
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
# Step 7: Train-test split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,   # 20% test data
    random_state=42  # for reproducibility
)

print("✂️ Data split done:")
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

```python
# Step 8: Train a RandomForestRegressor model

model = RandomForestRegressor(
    n_estimators=100,  # number of trees
    random_state=42
)

model.fit(X_train, y_train)

print("🚀 Model training complete!")
```

```python
# Step 9: Evaluate model performance

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("📊 MODEL EVALUATION METRICS")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE) : {mse:.4f}")
print(f"Root MSE (RMSE)          : {rmse:.4f}")
print(f"R² Score                 : {r2:.4f}")
```

```python
# Step 10: Plot actual vs predicted AQI for visual comparison

plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label="Actual AQI")
plt.plot(y_pred, label="Predicted AQI", alpha=0.7)
plt.title("Actual vs Predicted AQI")
plt.xlabel("Sample Index")
plt.ylabel("AQI")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Step 11: Define a function to map AQI values to categories (as per problem statement)

def aqi_category(aqi):
    """
    Map numerical AQI to category string.
    """
    if aqi <= 50:
        return "Good"
    elif aqi <= 100:
        return "Moderate"
    elif aqi <= 150:
        return "Unhealthy"
    elif aqi <= 200:
        return "Unhealthy for Strong People"
    else:
        return "Hazardous"

# Example:
for val in [30, 75, 120, 180, 250]:
    print(val, "->", aqi_category(val))
```

```python
# Apply category mapping on a few predictions

results_df = pd.DataFrame({
    "actual_aqi": y_test.values,
    "predicted_aqi": y_pred
})

# Take first 10 rows for display
results_df["actual_category"] = results_df["actual_aqi"].apply(aqi_category)
results_df["predicted_category"] = results_df["predicted_aqi"].apply(aqi_category)

results_df.head(10)
```

```python
# Step 12: Save the trained model to a file using joblib

import joblib

model_filename = "aqi_random_forest_model.pkl"
joblib.dump(model, model_filename)

print(f"💾 Model saved as {model_filename}")
```
